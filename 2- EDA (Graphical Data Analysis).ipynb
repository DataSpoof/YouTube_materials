# Exploratory Data analysis (EDA)

* Graphical data analysis
* Statistical data analysis


## Graphical data analysis

Univariate analysis (numerical or categorical)
  * numerical (histogram)
  * categorical (bar plot or count plot)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data_clean.csv")
df.head()
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head()
```

```python
df.shape
```

# Univariate analysis

```python
sns.histplot(df["Sales"])
plt.title("Distribution of Sales column")
plt.show()
```

```python
sns.countplot(x="ShelveLoc",data=df)
plt.title("Barplot of ShelveLoc column")
plt.show()
```

# Bivariate analysis (two columns at a time)
  * Two numerical column (scatter plot)
  * Two categorical column (stacked bar plot)
  * One numerical and one categorical column (histogram)

```python
plt.scatter(df['Sales'],df['Price'])
plt.title("Scatter plot of Sales and price column")
plt.show()
```

```python
sns.histplot(x="ShelveLoc",data=df,hue="US")
plt.title("Analyzing two categorical column")
plt.show()
```

```python
dfs= pd.melt(df,id_vars="Sales_category",value_vars=["Price"])

sns.barplot(x="Sales_category",y="value",data=dfs)
plt.title("Price wrt Sales category")
plt.show()
```

# Multivariate analysis
    
* more than two column (correlation)

```python
num_df=df.select_dtypes(include=['int','float'])
```

```python
cm=num_df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,cmap="coolwarm",fmt=".2f",square=True)
plt.title("correlation matrix")
plt.show()
```

```python
sns.pairplot(df)
```
